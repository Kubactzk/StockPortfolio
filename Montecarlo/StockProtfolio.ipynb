```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as pdr
import datetime as dt
```

```python
#import data
StockDir = "D:\Studia\ISA-magister\sem_1\Identyfikacja_i_modelowanie_statystyczne\Projekt\MonteCarlo\Tasla_Stock_Updated_V2.csv"
StockData = pd.read_csv(StockDir)
StockData.head()
# Date: The date of the stock record.
# Open: The opening price of Tesla's stock for the month.
# High: The highest price reached by Tesla's stock in the month.
# Low: The lowest price of Tesla's stock in the month.
# Close: The closing price of Tesla's stock for the month.
# Volume: The total trading volume of Tesla's stock for the month
```

```python
#Closing price data
data = StockData['Close']

#compute fractional change in previous stocks. Changes in days.
returns = data.pct_change()

mu, sigma = returns.mean(), returns.std()
initial = data.iloc[-1]
```

```python
for i in range(100):
    random = np.random.normal(mu,sigma,200)
    prices = initial * (random+1).cumprod()
    plt.axhline(initial, c='k')
    plt.plot(prices)
```
